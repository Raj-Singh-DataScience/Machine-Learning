# Predict canada's per capita income in year 2022 ?

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("D:\TestDirectory\ML\income.csv")  # read csv file
df
```

```python
plt.scatter(df.income,df.year)   # plottig all data points as scatter plot.
plt.xlabel("Income")
plt.ylabel("Year")
```

```python
df_new = df.drop('income', axis='columns')     #  dropped income in new data frame because we need year only as 2d object.
df_new
```

```python
model = linear_model.LinearRegression()           # creating model class object
model.fit(df_new,df.income)                       # using fit() method to train model
```

```python
model.predict([[2022]])    # Predict per capita income for 2022
```
